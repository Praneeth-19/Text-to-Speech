Internal Text

```python
from gtts import gTTS

text_to_say = "Hello!! Praneeth Burada. How was your day? And this Text is read by GTTS."

language = "en"

gtts_object = gTTS(text =  text_to_say,
                   lang = language,
                   slow = False)

gtts_object.save("/content/gtts.mp3")
```

```python
from IPython.display import Audio
Audio("/content/gtts.mp3", autoplay = True)
```

Changed Code (External Text)

```python
from gtts import gTTS
from IPython.display import Audio
```

```python
text_read = input("Enter the Text: ")
language = "en"

gtts_reader = gTTS(text =  text_read,
                   lang = language,
                   slow = False)
gtts_reader.save("/content/myread.mp3")

Audio("/content/myread.mp3", autoplay = True)
```
